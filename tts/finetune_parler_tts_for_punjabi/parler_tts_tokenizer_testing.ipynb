```python
from transformers import AutoTokenizer

# mT5 = multilingual T5, trained on 101 languages including Punjabi
tokenizer = AutoTokenizer.from_pretrained("google/mt5-small")

text = "ਸਤ ਸ੍ਰੀ ਅਕਾਲ ਮੇਰਾ ਨਾਮ ਅਮਨ ਹੈ"
tokens = tokenizer(text)

print("Token IDs:", tokens['input_ids'])
print("Number of tokens:", len(tokens['input_ids']))

# Most important test — decode back and see if Punjabi is preserved
decoded = tokenizer.decode(tokens['input_ids'], skip_special_tokens=True)
print("Decoded back:", decoded)
print("Match?", decoded.strip() == text.strip())
```
