```python
import librosa
import numpy as np
import pandas as pd
import os
from tqdm import tqdm

def analyze_audio(audio_path, text):
    y, sr = librosa.load(audio_path)

    # 1. Calculate Pitch
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    pitch = np.mean(pitches[pitches > 0]) if np.any(pitches > 0) else 0
    pitch_label = "low-pitched" if pitch < 150 else "high-pitched" if pitch > 250 else "moderate-pitched"

    # 2. Calculate Speaking Rate
    duration = librosa.get_duration(y=y, sr=sr)
    words = len(text.split())
    rate = words / duration
    rate_label = "slowly" if rate < 2 else "rapidly" if rate > 4 else "at a moderate speed"

    # 3. Simple Noise Check (SNR)
    rms = librosa.feature.rms(y=y)
    avg_rms = np.mean(rms)
    noise_label = "clear" if avg_rms > 0.02 else "slightly noisy"

    return f"A female speaker delivers this Punjabi text {rate_label} in a {pitch_label}, {noise_label} voice."

# Process your TSV
df = pd.read_csv('/content/train.tsv', sep='\t', header=None,
                 names=["id", "audio_path", "raw_transcription", "transcription", "phonetic", "duration", "gender"])

descriptions = []
for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    full_path = os.path.join("/content/wavs/", row['audio_path'])
    try:
        desc = analyze_audio(full_path, row['transcription'])
        descriptions.append(desc)
    except:
        descriptions.append("A female speaker delivers this Punjabi text.")

df['description'] = descriptions
df.to_csv("punjabi_final_data.csv", index=False)
print("вң… Done! Your data is ready in punjabi_final_data.csv")
```
