```python
import os
import IPython.display as ipd

# Configuration
text = "नमस्ते, मैं प्रियंवदा हूँ। मैं एक भारतीय आवाज़ हूँ जो बहुत तेज़ी से बोल सकती हूँ।"
model_filename = "hi_IN-priyamvada-medium.onnx"
output_filename = "priyamvada_output.wav"

# Get the absolute path to the uploaded file
# This is crucial for Piper to know and use my local file
model_path = os.path.abspath(model_filename)

# Run the Piper command-line tool
!echo "{text}" | piper --model {model_path} --output_file {output_filename}

# Verify and play the generated audio
if os.path.exists(output_filename):
    print("Speech generated successfully!")
    display(ipd.Audio(output_filename, autoplay=True))
else:
    print("Error: Audio file was not generated. Check if you uploaded the .onnx.json file as well.")
```
