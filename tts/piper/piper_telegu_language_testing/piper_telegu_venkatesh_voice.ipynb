```python
import os
import IPython.display as ipd

#for Telugu
text = "భారతదేశం అనేక భాషలు మరియు సంస్కృతులకు నిలయం, అందుకే ప్రతి ఒక్కరూ తమ మాతృభాషలో సాంకేతికతను సులభంగా ఉపయోగించుకునేలా చేయడం చాలా ముఖ్యం."
model_filename = "te_IN-venkatesh-medium.onnx"
output_filename = "venkatesh_telugu_output.wav"

# Get absolute path
model_path = os.path.abspath(model_filename)

# Run Piper for Telugu
!echo "{text}" | piper --model {model_path} --output_file {output_filename}

# Play audio
if os.path.exists(output_filename):
    print("Telugu speech generated successfully!")
    display(ipd.Audio(output_filename, autoplay=True))
else:
    print("Error: Audio file not generated. Ensure both te_IN files were downloaded.")
```
