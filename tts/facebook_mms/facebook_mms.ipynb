```python
from transformers import VitsModel, AutoTokenizer
import torch

model = VitsModel.from_pretrained("facebook/mms-tts-tel")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-tel")
```

```python
text = """వ్యవసాయంలో మంచి దిగుబడులు పొందాలంటే ముందుగా నేల పరీక్ష చేయించి మట్టి పోషక స్థితిని తెలుసుకోవాలి. \
పంటకు తగిన విత్తనాలను ఎంచుకుని ధృవీకరించిన విత్తనాలనే ఉపయోగించాలి.
విత్తనాలను విత్తే ముందు శుద్ధి చేసి విత్తన శుద్ధి మందులతో శుభ్రపరచడం మంచిది.
సమయానికి విత్తడం ద్వారా పంట వృద్ధి సక్రమంగా జరుగుతుంది.
నేలలో తేమను నిల్వ ఉంచేందుకు సేంద్రియ ఎరువులు మరియు పచ్చి ఎరువులు ఉపయోగించాలి."""
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform
```

```python
from IPython.display import Audio

Audio(output, rate=model.config.sampling_rate)
```

```python
import torchaudio

torchaudio.save("facebook_mms.wav", output, model.config.sampling_rate)
print('facebook_mms.wav saved..!!')
```
