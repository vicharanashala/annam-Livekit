```python
import requests
import json

# ========== XLIT CONFIG ==========
BASE_URL = "https://mesne-unlicentiously-allie.ngrok-free.dev/xlit"

SUPPORTED_LANGS = [
    "as", "bn", "brx", "en", "gom", "gu", "hi", "kn", "ks",
    "mai", "ml", "mni", "mr", "ne", "or", "pa", "sa",
    "sd", "si", "ta", "te", "ur"
]

def transliterate_en_to_indic(text, target_languages, beam_width=4):
    invalid = [lang for lang in target_languages if lang not in SUPPORTED_LANGS]
    if invalid:
        raise ValueError(f"Unsupported language(s): {invalid}")

    payload = {
        "text": text,
        "target_languages": target_languages,
        "beam_width": beam_width
    }

    response = requests.post(
        f"{BASE_URL}/transliterate/en-to-indic",
        json=payload
    )

    response.raise_for_status()
    return response.json()


# ========== TTS FUNCTION ==========
def generate_tts_wav(text: str, ref_audio_path: str, ref_text: str, output_file: str = "final.wav"):
    url = "https://mesne-unlicentiously-allie.ngrok-free.dev/tts_wav"

    data = {
        "text": text,
        "ref_text": ref_text
    }

    files = {
        "ref_audio_file": open(ref_audio_path, "rb")
    }

    try:
        response = requests.post(url, data=data, files=files)

        if response.status_code == 200:
            with open(output_file, "wb") as f:
                f.write(response.content)
            print(f"✅ Audio saved as {output_file}")
        else:
            print("❌ Request failed:", response.status_code)
            print(response.text)

    finally:
        files["ref_audio_file"].close()


# ========== MAIN ==========
raw_text = "aaji prashnottara system re piecha maan 5.5 gunita 10 ru ghaata minus 3 molar jaancha kara gala"

#Call IndicXlit for Bengali
result = transliterate_en_to_indic(
    text=raw_text,
    target_languages=["or"]
)

print("XLIT RESPONSE:")
print(json.dumps(result, indent=2, ensure_ascii=False))

processed_text = result["results"]["or"]

print("\nOriginal text:", raw_text)
print("Processed text:", processed_text)

#Call TTS
generate_tts_wav(
    text=processed_text,
    ref_audio_path="/content/oriya_voice (1).wav",
    ref_text="ନଦୀ ପାଖରେ ଗଛତଳେ ବସି ମୁଁ ସକାଳର ସୁନ୍ଦର ପବନକୁ ଉପଭୋଗ କଲି।",
    output_file="final.wav"
)
```

# **Example 5(with xlit and regex normalization)**

```python
import requests
import json
import re

# ========== XLIT CONFIG ==========
BASE_URL = "https://mesne-unlicentiously-allie.ngrok-free.dev/xlit"

SUPPORTED_LANGS = [
    "as", "bn", "brx", "en", "gom", "gu", "hi", "kn", "ks",
    "mai", "ml", "mni", "mr", "ne", "or", "pa", "sa",
    "sd", "si", "ta", "te", "ur"
]

def transliterate_en_to_indic(text, target_languages, beam_width=4):
    invalid = [lang for lang in target_languages if lang not in SUPPORTED_LANGS]
    if invalid:
        raise ValueError(f"Unsupported language(s): {invalid}")

    payload = {
        "text": text,
        "target_languages": target_languages,
        "beam_width": beam_width
    }

    response = requests.post(
        f"{BASE_URL}/transliterate/en-to-indic",
        json=payload
    )

    response.raise_for_status()
    return response.json()


# ========== PRE-NORMALIZE TOKENS ==========
def normalize_tokens(text: str) -> str:
    # Help XLIT pronounce these better
    text = text.replace("pH", "pee ech")
    text = text.replace("FAQ", "F A Q")
    return text


# ========== SCIENTIFIC NOTATION NORMALIZER (ODIA) ==========

DIGIT_MAP = {
    '0': 'ଶୂନ୍ୟ', '1': 'ଏକ', '2': 'ଦୁଇ', '3': 'ତିନି', '4': 'ଚାରି',
    '5': 'ପାଞ୍ଚ', '6': 'ଛଅ', '7': 'ସାତ', '8': 'ଆଠ', '9': 'ନଅ'
}

SUPERSCRIPT_MAP = {
    '⁰': '0', '¹': '1', '²': '2', '³': '3', '⁴': '4',
    '⁵': '5', '⁶': '6', '⁷': '7', '⁸': '8', '⁹': '9',
    '⁻': '-'
}

def superscript_to_normal(s):
    return "".join(SUPERSCRIPT_MAP.get(ch, ch) for ch in s)

def speak_number(num):
    if '.' in num:
        a, b = num.split('.')
        return DIGIT_MAP[a] + " ଦଶମିକ " + " ".join(DIGIT_MAP[d] for d in b)
    else:
        return " ".join(DIGIT_MAP[d] for d in num)

def speak_power(p):
    p = int(p)
    if p < 0:
        return "ମାଇନସ୍ " + " ".join(DIGIT_MAP[d] for d in str(abs(p)))
    else:
        return " ".join(DIGIT_MAP[d] for d in str(p))

# FIXED REGEX: supports x, ×, ଏକ୍ସ
SCI_PATTERN = re.compile(
    r'(\d+(?:\.\d+)?)\s*(x|×|ଏକ୍ସ)\s*10\^?([⁻⁰¹²³⁴⁵⁶⁷⁸⁹\-0-9]+)'
)

def normalize_scientific_notation(text: str) -> str:
    def replacer(match):
        number = match.group(1)
        power_raw = match.group(3)

        power_norm = superscript_to_normal(power_raw)

        spoken_num = speak_number(number)
        spoken_pow = speak_power(power_norm)

        return f"{spoken_num} ଗୁଣିତ ଦଶ ରୁ ଘାତ {spoken_pow}"

    return SCI_PATTERN.sub(replacer, text)


# ========== TTS FUNCTION ==========
def generate_tts_wav(text: str, ref_audio_path: str, ref_text: str, output_file: str = "final.wav"):
    url = "https://mesne-unlicentiously-allie.ngrok-free.dev/tts_wav"

    data = {
        "text": text,
        "ref_text": ref_text
    }

    files = {
        "ref_audio_file": open(ref_audio_path, "rb")
    }

    try:
        response = requests.post(url, data=data, files=files)

        if response.status_code == 200:
            with open(output_file, "wb") as f:
                f.write(response.content)
            print(f"✅ Audio saved as {output_file}")
        else:
            print("❌ Request failed:", response.status_code)
            print(response.text)

    finally:
        files["ref_audio_file"].close()


# ========== MAIN ==========
raw_text = "aaji lab re pH value 5.5 x 10^-3 note kara gala ebang taku FAQ page re deya gala"

# Pre-normalize tokens
raw_text = normalize_tokens(raw_text)

# Call IndicXlit for Odia
result = transliterate_en_to_indic(
    text=raw_text,
    target_languages=["or"]
)

print("XLIT RESPONSE:")
print(json.dumps(result, indent=2, ensure_ascii=False))

processed_text = result["results"]["or"]

print("\nAfter XLIT:", processed_text)

# Normalize scientific notation
processed_text = normalize_scientific_notation(processed_text)

print("\nAfter regex normalization:", processed_text)

# Call TTS
generate_tts_wav(
    text=processed_text,
    ref_audio_path="/content/oriya_voice (1).wav",
    ref_text="ନଦୀ ପାଖରେ ଗଛତଳେ ବସି ମୁଁ ସକାଳର ସୁନ୍ଦର ପବନକୁ ଉପଭୋଗ କଲି।",
    output_file="final.wav"
)
```
