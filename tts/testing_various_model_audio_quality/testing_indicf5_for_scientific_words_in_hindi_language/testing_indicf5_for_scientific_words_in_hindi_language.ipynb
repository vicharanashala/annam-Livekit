```python
import requests

def generate_tts_wav(text: str, ref_audio_path: str, ref_text: str, output_file: str = "upload_test.wav"):
    url = "https://mesne-unlicentiously-allie.ngrok-free.dev/tts_wav"

    # Form fields
    data = {
        "text": text,
        "ref_text": ref_text
    }

    # File field
    files = {
        "ref_audio_file": open(ref_audio_path, "rb")
    }

    try:
        response = requests.post(url, data=data, files=files)

        if response.status_code == 200:
            with open(output_file, "wb") as f:
                f.write(response.content)
            print(f"Audio saved as {output_file}")
        else:
            print("Request failed:", response.status_code)
            print(response.text)

    finally:
        files["ref_audio_file"].close()
```

# **With Xlit**

```python
import requests
import json

# ========== XLIT CONFIG ==========
BASE_URL = "https://mesne-unlicentiously-allie.ngrok-free.dev/xlit"

SUPPORTED_LANGS = [
    "as", "bn", "brx", "en", "gom", "gu", "hi", "kn", "ks",
    "mai", "ml", "mni", "mr", "ne", "or", "pa", "sa",
    "sd", "si", "ta", "te", "ur"
]

def transliterate_en_to_indic(text, target_languages, beam_width=4):
    invalid = [lang for lang in target_languages if lang not in SUPPORTED_LANGS]
    if invalid:
        raise ValueError(f"Unsupported language(s): {invalid}")

    payload = {
        "text": text,
        "target_languages": target_languages,
        "beam_width": beam_width
    }

    response = requests.post(
        f"{BASE_URL}/transliterate/en-to-indic",
        json=payload
    )

    response.raise_for_status()
    return response.json()


# ========== TTS FUNCTION ==========
def generate_tts_wav(text: str, ref_audio_path: str, ref_text: str, output_file: str = "final.wav"):
    url = "https://mesne-unlicentiously-allie.ngrok-free.dev/tts_wav"

    data = {
        "text": text,
        "ref_text": ref_text
    }

    files = {
        "ref_audio_file": open(ref_audio_path, "rb")
    }

    try:
        response = requests.post(url, data=data, files=files)

        if response.status_code == 200:
            with open(output_file, "wb") as f:
                f.write(response.content)
            print(f"✅ Audio saved as {output_file}")
        else:
            print("❌ Request failed:", response.status_code)
            print(response.text)

    finally:
        files["ref_audio_file"].close()


# ========== MAIN ==========
raw_text = "aaj FAQ system mein pH value check ki gayi hai"

#Call IndicXlit for Bengali
result = transliterate_en_to_indic(
    text=raw_text,
    target_languages=["hi"]
)

print("XLIT RESPONSE:")
print(json.dumps(result, indent=2, ensure_ascii=False))

processed_text = result["results"]["hi"]

print("\nOriginal text:", raw_text)
print("Processed text:", processed_text)

#Call TTS
generate_tts_wav(
    text=processed_text,
    ref_audio_path="/content/Hindi_F_Deepika.wav",
    ref_text="आगरा के ताज महल में मुगल बादशाह शाहजहाँ के तीन सौ साठ में उर्स की शुरुआत की गई, जिसके चलते शाहजहाँ और मुमताज़ की कब्र का गेट खोला गया।",
    output_file="final.wav"
)
```
